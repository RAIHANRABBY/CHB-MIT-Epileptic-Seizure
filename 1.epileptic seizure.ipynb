```python
import os
import mne
import pywt
import numpy as np

# Load and Parse Data
def load_and_parse_data(P_folder):
    all_edf_file_paths = [os.path.join(P_folder, f) for f in os.listdir(P_folder) if f.endswith('.edf')]
    summary_file_path = [os.path.join(P_folder, f) for f in os.listdir(P_folder) if f.endswith('.txt')][0]
    
    with open(summary_file_path, 'r') as file:
        summary_content = file.readlines()

    seizure_times = parse_seizure_times(summary_content)
    return all_edf_file_paths, seizure_times

def parse_seizure_times(summary_content):
    seizure_times = {}
    current_file = None
    for line in summary_content:
        if 'File Name' in line:
            current_file = line.split(':')[-1].strip()
            seizure_times[current_file] = []
        elif 'Seizure Start Time' in line:
            start_time = int(line.split(':')[-1].strip().split(' ')[0])
        elif 'Seizure End Time' in line:
            end_time = int(line.split(':')[-1].strip().split(' ')[0])
            seizure_times[current_file].append((start_time, end_time))
    return seizure_times

# Annotate EEG Data
def annotate_eeg_data(raw, seizure_times, preictal_duration=30*60, postictal_duration=10*60, buffer_duration=30*60):
    file_name = os.path.basename(raw.filenames[0])

    if file_name in seizure_times:
        for start, end in seizure_times[file_name]:
            # Exclude buffer before preictal period
            if start - preictal_duration - buffer_duration > 0:
                raw.annotations.append(onset=start - preictal_duration - buffer_duration, duration=buffer_duration, description='buffer_before_preictal')
            
            # Annotate preictal period
            preictal_start = max(0, start - preictal_duration)
            preictal_duration_actual = start - preictal_start
            raw.annotations.append(onset=preictal_start, duration=preictal_duration_actual, description='preictal')
            
            # Annotate ictal period
            duration = end - start
            raw.annotations.append(onset=start, duration=duration, description='ictal')
            
            # Annotate postictal period
            postictal_start = end
            postictal_duration_actual = min(postictal_duration, raw.times[-1] - postictal_start)
            raw.annotations.append(onset=postictal_start, duration=postictal_duration_actual, description='postictal')
            
            # Exclude buffer after postictal period
            if postictal_start + postictal_duration_actual + buffer_duration < raw.times[-1]:
                raw.annotations.append(onset=postictal_start + postictal_duration_actual, duration=buffer_duration, description='buffer_after_postictal')
            
        # Annotate remaining time as interictal
        last_end = 0
        for start, end in seizure_times[file_name]:
            if last_end < preictal_start:
                raw.annotations.append(onset=last_end, duration=preictal_start - last_end, description='interictal')
            last_end = end + postictal_duration_actual + buffer_duration
        if last_end < raw.times[-1]:
            raw.annotations.append(onset=last_end, duration=raw.times[-1] - last_end, description='interictal')
    else:
        raw.annotations.append(onset=0, duration=raw.times[-1], description='interictal')

    return raw

# Preprocess the EEG Data
def preprocess_eeg_data(raw):
    raw = ChannelSelection(raw)
    raw.filter(l_freq=0.5, h_freq=127.9)
    raw_notch = raw.copy().notch_filter(freqs=[60, 120], notch_widths=6)
    return raw_notch

def ChannelSelection(signal):
    try:
        channels = [
            'FP1-F7', 'F7-T7', 'T7-P7', 'P7-O1',
            'FP1-F3', 'F3-C3', 'C3-P3', 'P3-O1',
            'FP2-F4', 'F4-C4', 'C4-P4', 'P4-O2',
            'FP2-F8', 'F8-T8', 'T8-P8-0', 'P8-O2',
            'FZ-CZ', 'CZ-PZ', 'T7-FT9',
            'FT9-FT10', 'FT10-T8',
        ]

        signal.pick_channels(channels)
        channel_mapping = {'T8-P8-0': 'T8-P8'}
        mne.rename_channels(signal.info, channel_mapping)
        return signal
    except:
        print("Channel names are not the same!!")
        return signal

def DWT(data):
    eeg_data = data.get_data()
    dwt_coeffs = []
    for channel_data in eeg_data:
        wavelet = 'db4'
        coeffs = pywt.wavedec(channel_data, wavelet, level=5)
        dwt_coeffs.append(coeffs)
    return dwt_coeffs

# Create Epochs and Labels
def create_epochs_and_labels(raw, epoch_length=10):
    epoch_samples = int(epoch_length * raw.info['sfreq'])

    # Predefined possible event ID mapping for consistency across all files
    event_ids = {
        'ictal': 1,
        'preictal': 2,
        'postictal': 3,
        'interictal': 4
    }

    # Filter out buffer annotations before creating events
    annotations = [ann for ann in raw.annotations if ann['description'] not in ['buffer_before_preictal', 'buffer_after_postictal']]
    
    # Only keep event IDs for annotations that are present in the current file
    present_event_ids = {desc: event_ids[desc] for desc in event_ids if desc in [ann['description'] for ann in annotations]}
    
    # Create events based on annotations
    events, _ = mne.events_from_annotations(raw, event_id=present_event_ids)

    # Debugging information
    print(f"Events array: {events}")
    print(f"Present Event IDs: {present_event_ids}")
    
    # Create fixed-length epochs
    epochs = mne.make_fixed_length_epochs(raw, duration=epoch_length, preload=True, overlap=0)

    # Ensure correct IDs are used in Y
    Y = np.array([present_event_ids.get(epochs.event_id[id], -1) for id in epochs.events[:, -1]])

    # Debugging to verify if interictal periods are correctly identified
    print("Epochs Events: ", epochs.events)
    print("Labels (Y): ", Y)
    print("------------------------------------------------------------------------------------------------------------------------\n",epochs.events)
    return epochs.get_data(), Y, epochs.events, epochs

    


# Combine Methods to Process and Annotate Data
def process_and_annotate(P_folder):
    all_edf_file_paths, seizure_times = load_and_parse_data(P_folder)
    X_all, Y_all = [], []

    for edf_file in all_edf_file_paths[:5]:
        print("###############################################",edf_file,"###########################################################")
        raw = mne.io.read_raw_edf(edf_file, preload=True)
        raw = annotate_eeg_data(raw, seizure_times)
        raw = preprocess_eeg_data(raw)

        X, Y , events , epochs= create_epochs_and_labels(raw)
        X_all.append(X)
        Y_all.append(Y)
        break

    # Stack the arrays vertically using np.vstack
    # X_all = np.vstack(X_all)
    # Y_all = np.hstack(Y_all)  # hstack is used to concatenate the labels horizontally
    
    return X_all, Y_all, events, epochs

# Example usage
P_folder = 'D:\\CHB_MIT'
X, Y , events, epochs= process_and_annotate(P_folder)
```

```python
def set_reference(data):
    # List of EEG channels (based on the names you provided)
    eeg_channels = ['EEG Fp1', 'EEG F3', 'EEG C3', 'EEG P3', 'EEG O1', 'EEG F7', 
                'EEG T3', 'EEG T5', 'EEG Fc1', 'EEG Fc5', 'EEG Cp1', 'EEG Cp5', 
                'EEG F9', 'EEG Fz', 'EEG Cz', 'EEG Pz', 'EEG Fp2', 'EEG F4', 
                'EEG C4', 'EEG P4', 'EEG O2', 'EEG F8', 'EEG T4', 'EEG T6', 
                'EEG Fc2', 'EEG Fc6', 'EEG Cp2', 'EEG Cp6', 'EEG F10']

    # Apply the average reference
    data.set_eeg_reference(ref_channels='average', projection=False, ch_type='eeg')


    # Set the channel types
    data.set_channel_types({'EKG EKG': 'ecg', 'SPO2': 'misc', 'HR': 'misc', '1': 'misc', '2': 'misc', 'MK': 'misc'})
    return data



# setting montage funciton
def SetMontage(data):
    montage = mne.channels.make_standard_montage('standard_1020')
    rename_dict={'EEG Fp1': 'Fp1',
 'EEG F3': 'F3',
 'EEG C3': 'C3',
 'EEG P3': 'P3',
 'EEG O1': 'O1',
 'EEG F7': 'F7',
 'EEG T3': 'T3',
 'EEG T5': 'T5',
 'EEG Fc1': 'FC1',
 'EEG Fc5': 'FC5',
 'EEG Cp1': 'CP1',
 'EEG Cp5': 'CP5',
 'EEG F9': 'F9',
 'EEG Fz': 'Fz',
 'EEG Cz': 'Cz',
 'EEG Pz': 'Pz',
 'EEG Fp2': 'Fp2',
 'EEG F4': 'F4',
 'EEG C4': 'C4',
 'EEG P4': 'P4',
 'EEG O2': 'O2',
 'EEG F8': 'F8',
 'EEG T4': 'T4',
 'EEG T6': 'T6',
 'EEG Fc2': 'FC2',
 'EEG Fc6': 'FC6',
 'EEG Cp2': 'CP2',
 'EEG Cp6': 'CP6',
 'EEG F10': 'F10'}
    data.rename_channels(rename_dict)
    data.set_montage(montage)

    return data
```

```python
import os
import mne
import numpy as np

# Define the directory and file paths
data_dir = 'D:\\CHB_MIT'
output_dir = os.path.join(data_dir, 'preprocessed')
os.makedirs(output_dir, exist_ok=True)

all_edf_file_paths = [os.path.join(data_dir, f) for f in os.listdir(data_dir) if f.endswith('.edf')]
summary_file_path = [os.path.join(data_dir, f) for f in os.listdir(data_dir) if f.endswith('summary.txt')][0]

# Load the summary file
with open(summary_file_path, 'r') as file:
    summary_content = file.readlines()

# Parse seizure times from the summary file
def parse_seizure_times(summary_content):
    seizure_times = {}
    current_file = None
    for line in summary_content:
        if 'File Name' in line:
            current_file = line.split(':')[-1].strip()
            seizure_times[current_file] = []
        elif 'Seizure Start Time' in line:
            start_time = int(line.split(':')[-1].strip())
        elif 'Seizure End Time' in line:
            end_time = int(line.split(':')[-1].strip())
            seizure_times[current_file].append((start_time, end_time))
    return seizure_times

seizure_times = parse_seizure_times(summary_content)

# Function to preprocess each EDF file
def preprocess_edf(edf_file, seizure_times, output_dir):
    raw = mne.io.read_raw_edf(edf_file, preload=True)
    file_name = os.path.basename(edf_file)
    
    if file_name in seizure_times:
        for start, end in seizure_times[file_name]:
            # Annotate seizures
            onset = start
            duration = (end - start)
            raw.annotations.append(onset=onset, duration=duration, description='seizure')
    
    # Set montage
    montage = mne.channels.make_standard_montage('standard_1020')
    raw.set_montage(montage)
    
    # Apply band-pass filter
    raw.filter(0.5, 50, fir_design='firwin')
    
    # Apply notch filter to remove power line noise
    raw.notch_filter(freqs=[50], fir_design='firwin')
    
    # Identify and remove artifacts using ICA
    ica = mne.preprocessing.ICA(n_components=20, random_state=97, max_iter=800)
    ica.fit(raw)
    ica.plot_components()
    
    # Manually inspect the components and update this list
    ica.exclude = [0, 1]  # Example: components to exclude, update based on visual inspection
    ica.apply(raw)
    
    # Create epochs around seizures
    events = mne.find_events(raw, stim_channel='STI 014')
    event_id = {'seizure': 1}
    epochs = mne.Epochs(raw, events, event_id=event_id, tmin=-0.2, tmax=0.5, baseline=(None, 0))
    
    # Resample if necessary
    raw.resample(256, npad="auto")
    
    # Standardize the data
    data = epochs.get_data()
    data -= np.mean(data, axis=0)
    data /= np.std(data, axis=0)
    
    # Save the preprocessed data
    save_path = os.path.join(output_dir, f'{os.path.basename(raw.filenames[0]).split(".")[0]}-epo.fif')
    epochs.save(save_path, overwrite=True)
    return raw

# Process each EDF file
for edf_file in all_edf_file_paths:
    preprocess_edf(edf_file, seizure_times, output_dir)
```

```python
import pandas as pd
import re

def extract_all_annotation_info(file_path):
    annotations = []
    with open(file_path, 'r') as f:
        content = f.read()
    blocks = content.split("File name:")
    blocks = blocks[1:]  # Exclude the first split part as it will be empty
    for block in blocks:
        info = {}
        
        match = re.search(r'(\S+)', block)
        if match:
            info['File name'] = match.group(1)
        
        match = re.search(r'Registration start time:\s*(\S+)', block)
        if match:
            info['Registration start time'] = match.group(1)
        
        match = re.search(r'Registration end time:\s*(\S+)', block)
        
        if match:
            info['Registration end time'] = match.group(1)
        
        match = re.search(r'(Seizure start time:|Start time:)\s*(\S+)', block)
        if match:
            info['Seizure start time'] = match.group(2)
        
        match = re.search(r'(Seizure end time:|End time:)\s*(\S+)', block)
        if match:
            info['Seizure end time'] = match.group(2)
            
        annotations.append(info)
    return annotations

# Extracting data from the .txt file
annotations = extract_all_annotation_info('D:\\Research\\EEG Datasets\\siena-scalp-eeg-database-1.0.0\\PN00\\Seizures-list-PN00.txt')

# Converting the list of dictionaries to a DataFrame
df = pd.DataFrame(annotations)

# Displaying the DataFrame
df
```

```python
def set_reference(data):
    # List of EEG channels (based on the names you provided)
    eeg_channels = ['EEG Fp1', 'EEG F3', 'EEG C3', 'EEG P3', 'EEG O1', 'EEG F7', 
                'EEG T3', 'EEG T5', 'EEG Fc1', 'EEG Fc5', 'EEG Cp1', 'EEG Cp5', 
                'EEG F9', 'EEG Fz', 'EEG Cz', 'EEG Pz', 'EEG Fp2', 'EEG F4', 
                'EEG C4', 'EEG P4', 'EEG O2', 'EEG F8', 'EEG T4', 'EEG T6', 
                'EEG Fc2', 'EEG Fc6', 'EEG Cp2', 'EEG Cp6', 'EEG F10']

    # Apply the average reference
    data.set_eeg_reference(ref_channels='average', projection=False, ch_type='eeg')


    # Set the channel types
    data.set_channel_types({'EKG EKG': 'ecg', 'SPO2': 'misc', 'HR': 'misc', '1': 'misc', '2': 'misc', 'MK': 'misc'})
    return data

def SetMontage(data):
    montage = mne.channels.make_standard_montage('standard_1020')
    rename_dict={'EEG Fp1': 'Fp1',
 'EEG F3': 'F3',
 'EEG C3': 'C3',
 'EEG P3': 'P3',
 'EEG O1': 'O1',
 'EEG F7': 'F7',
 'EEG T3': 'T3',
 'EEG T5': 'T5',
 'EEG Fc1': 'FC1',
 'EEG Fc5': 'FC5',
 'EEG Cp1': 'CP1',
 'EEG Cp5': 'CP5',
 'EEG F9': 'F9',
 'EEG Fz': 'Fz',
 'EEG Cz': 'Cz',
 'EEG Pz': 'Pz',
 'EEG Fp2': 'Fp2',
 'EEG F4': 'F4',
 'EEG C4': 'C4',
 'EEG P4': 'P4',
 'EEG O2': 'O2',
 'EEG F8': 'F8',
 'EEG T4': 'T4',
 'EEG T6': 'T6',
 'EEG Fc2': 'FC2',
 'EEG Fc6': 'FC6',
 'EEG Cp2': 'CP2',
 'EEG Cp6': 'CP6',
 'EEG F10': 'F10'}
    data.rename_channels(rename_dict)
    data.set_montage(montage)

    return data


def apply_ica(data):
    ica = mne.preprocessing.ICA(n_components=0.95, random_state=42, max_iter=800)
    ica.fit(data)
    ecg_inds, scores = ica.find_bads_ecg(data, ch_name='EKG EKG',threshold= 0.5)
    ica.exclude = ecg_inds
    raw_ica = ica.apply(data.copy(), exclude=ica.exclude)
    return raw_ica
    




def data_load(path):
    if isinstance(path,str) :
        data = mne.io.read_raw_edf(path,preload= True)
    else:
        data= path
    data_ref = set_reference(data)
    data_mont = SetMontage(data_ref)
    ica_data = apply_ica(data_mont)

    return ica_data

    
```

## Mark the events and epoching

```python
def time_to_seconds(time_str):
        """Converts time from "HH.MM.SS" format to seconds."""
        hours, minutes, seconds = map(int, time_str.split('.'))
        return hours * 3600 + minutes * 60 + seconds

def annotate_seizure(data, reg_start_time, seizure_start_time, seizure_end_time):
    """
    Annotates the seizure events on the EEG raw data.

    Parameters:
    - raw: mne.io.Raw object containing EEG data.
    - reg_start_time: str, registration start time in "HH.MM.SS" format.
    - seizure_start_time: str, seizure start time in "HH.MM.SS" format.
    - seizure_end_time: str, seizure end time in "HH.MM.SS" format.
    """

    

    sfreq = data.info['sfreq']

    # Calculate relative seizure start and end times in seconds
    relative_seizure_start = time_to_seconds(seizure_start_time) - time_to_seconds(reg_start_time)
    relative_seizure_end = time_to_seconds(seizure_end_time) - time_to_seconds(reg_start_time)

    # Convert relative times to samples
    start_sample = int(relative_seizure_start * sfreq)
    end_sample = int(relative_seizure_end * sfreq)

    # Annotate the seizure
    annotations = mne.Annotations(onset=start_sample/sfreq, duration=(end_sample-start_sample)/sfreq, description='Seizure')
    data.set_annotations(annotations)

    # Return the annotated raw object
    return data


```

```python
# Calculate the number of segments
n_segments = int(raw.times[-1] // segment_length_sec)
n_segments
```

```python

# Save each segment into a new file
for i in range(n_segments):
    start_sample = i * segment_length_sec * raw.info['sfreq']
    stop_sample = (i + 1) * segment_length_sec * raw.info['sfreq']
    
    # Crop the raw data to the current segment
    raw_segment = raw.copy().crop(tmin=start_sample / raw.info['sfreq'], tmax=min(stop_sample / raw.info['sfreq'], raw.times[-1]))
    
    # Check if the annotated event falls within this segment
    if onset >= start_sample and onset + duration <= stop_sample:
        # Add the annotation to this segment
        raw_segment.set_annotations(mne.Annotations(onset=[onset - start_sample], duration=[duration], description=[description]))

    
    raw_segment_ref = set_reference(raw_segment)
    raw_segment_mont = SetMontage(raw_segment_ref)
    raw_segment_ica = apply_ica(raw_segment_mont)
    # Save the segment
    raw_segment_ica.save(f'D:\\Research\\EEG Datasets\\\evented_epileptic_data\\PN03\\PN03-2\\raw_pn03_{i}.fif', overwrite=True)
```

## epoch the data

```python
raw.ch_names
```

```python
events = mne.find_events(raw, stim_channel='2')
events
```

```python

# Define the time range around the event to extract: -0.5 to 1.0 sec around the event
tmin = -10.0
tmax = 10.0

# Create epochs
epochs = mne.Epochs(raw, events, event_id=None, tmin=tmin, tmax=tmax, baseline=None, preload=True)

# Optionally, visualize the epochs
epochs.plot()
```

```python
# Create an evoked response by averaging epochs
evoked = epochs.average()

# Plot the evoked response
evoked.plot()
```

```python
times = np.arange(-7.0,10.0,1)
fig = evoked.plot_topomap(times=times)
fig.suptitle("Standard")
```
